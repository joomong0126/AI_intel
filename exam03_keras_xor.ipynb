```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
training_data = np.array(
[[0,0],[0,1],[1,0],[1,1]],'float32' )
target_data=np.array([[0],[1],[1],[0]],'float32') # 0,0 은 0 0,1은 1 이런식
```

```python
model = Sequential()  # model 만들기
model.add(Dense(32,input_dim=2,activation='relu')) # nput_dim=2 입력이 2개이기 때문에, relu
model.add(Dense(1,activation='sigmoid')) # sigmoid 를 씀
model.compile(loss='mse',optimizer='adam',metrics=['binary_accuracy']) # optimizer 경사각 알고리즘은 adam 을 사용함. ['binary_accuracy'] 이진 분류를 함.  Adam은 매우 효과적이면서 널리 사용되는 최적화 알고리즘
model.summary()
```

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
inp = list(map(int,input().split())) # 사용자로부터 입력을 받아 정수 리스트로 변환
qwe= np.array(inp) # NumPy 배열로 변환
print('입력값')
qwe= qwe.reshape(1,2) # 1x2 차원으로 변형
print('결과값',model.predict(qwe)[0][0].round()) # round() 함수를 사용해서 반올림을 씀.
# 훈련을 통해서 weight 값을 키울 수 있어. 이걸 통해서 1로 나오게 할 수 있어.
```
