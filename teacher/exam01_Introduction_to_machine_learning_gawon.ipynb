#머신러닝(딥러닝) 입문

##Linear regression

###기존의 프로그램 방식

```python
def celsius_to_fahrenheit(x): # 섭씨온도를 화씨 온도로 바꾸는 작업
    return x * 1.8 + 32
```

```python
celsius_value = int(input('섭씨온도를 입력하세요.'))
print('화씨온도로', celsius_to_fahrenheit(celsius_value))
```

###머신러닝 방식

```python
from tensorflow.keras.models import Sequential #tensorflow.keras.models 패키지에서 Sequential 모듈을 가져옴
from tensorflow.keras.layers import Dense, InputLayer #tensorflow.keras.layers 패키지에서 Dense, InputLayer 모듈을 가져옴
import numpy as np # numpy 모듈의 이름을 np로 지정하기
import matplotlib.pyplot as plt   #  matplotlib.pyplot 모듈의 이름을 plt로 지정하기
```

```python
data_C = np.array(range(0, 100)) # 0부터 100까지 리스트로 집어 넣기
data_F = celsius_to_fahrenheit(data_C) # celsius_to_fahrenheit 을 이용해서 화씨 온도로 바꾸기
print(data_C)
print(data_F)
```

```python
model = Sequential() # Sequential 모델은 계층을 선형으로 쌓은 것입니다.
model.add(InputLayer(input_shape=(1,)))
model.add(Dense(1)) # Dense의 첫번째 파라미터는 노드 수를 의미한다.
model.compile(loss='mse', optimizer='rmsprop') # compile() 함수로 뉴럴렛을 학습시키기 위한 optimizer와 loss function을 정의할 수 있다.
model.summary()
```

```python
scaled_data_C = data_C / 100
scaled_data_F = data_F / 100
print(scaled_data_C)
print(scaled_data_F)
```

```python
print(model.predict([0.01]))
```

```python
model.save('before_learning.h5')
```

```python
fit_hist = model.fit(scaled_data_C, scaled_data_F, epochs=1000) # epochs 는 트레이닝 데이터를 몇 번 사용해서 학습 할 것인지에 대한 변수
```

```python
print(model.predict([0.01]))
```

```python
model.save('after_learning.h5')
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
noise = np.array(np.random.normal(0, 0.05, 100))
print(noise)
```

```python
noised_scaled_data_F = np.array([])
for data in scaled_data_F:
    noised_scaled_data_F = np.append(
        noised_scaled_data_F, np.random.normal(0, 0.05, 100) + data)
print(noised_scaled_data_F)
print(len(noised_scaled_data_F))
```

```python
noised_scaled_data_C = []
for data in range(0, 100):
    for i in range(0, 100):
        noised_scaled_data_C.append(data)
noised_scaled_data_C = np.array(noised_scaled_data_C)
noised_scaled_data_C = noised_scaled_data_C / 100
print(noised_scaled_data_C)
print(len(noised_scaled_data_C))
```

```python
plt.scatter(x=noised_scaled_data_C, y=noised_scaled_data_F)
plt.show()
```

```python
fig = plt.figure(figsize=(50, 50))
ax = fig.add_subplot(111)
ax.scatter(x=noised_scaled_data_C, y=noised_scaled_data_F,
           alpha=0.2, s=200, marker='+')
plt.show()
```

```python
model2 = Sequential()
model2.add(InputLayer(input_shape=(1,)))
model2.add(Dense(1))
model2.compile(loss='mse', optimizer='rmsprop')
model2.summary()
```

```python
model2.predict([0.01])
```

```python
fit_hist = model2.fit(noised_scaled_data_C, noised_scaled_data_F,
                      epochs=20)
```

```python
model2.save('noised_after_learning.h5')
```

```python
print(model2.predict([0.01]))
```

```python
celsius_value = int(input('섭씨 온도를 입력하세요.'))
print('화씨온도로',
      np.around(model2.predict([celsius_value / 100]) * 100, 1), '입니다.')
```
