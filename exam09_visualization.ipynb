```python
cmaps = {}

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))


def plot_color_gradients(category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axs = plt.subplots(nrows=nrows + 1, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh,
                        left=0.2, right=0.99)
    axs[0].set_title(f'{category} colormaps', fontsize=14)

    for ax, name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=mpl.colormaps[name])
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)
# va : 수직 정렬, ha: 수평정렬

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()

    # Save colormap list for later.
    cmaps[category] = cmap_list
```

```python
# x = np.arrange(n) # 리스트를 대신해서 쓰면 편함.
import numpy as np
import matplotlib.pyplot as plt

# 데이터 포인트 개수
n = 5

# x 값을 0부터 n-1까지의 정수 리스트로 생성
ind = np.arange(n)

# 남성과 여성의 평균 점수 및 표준 편차
menMeans = (20, 35, 30, 35, -27)
womenMeans = (25, 32, 34, 20, -25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)

# 막대의 너비
width = 0.35

# 새로운 그림과 축 생성
fig, ax = plt.subplots()

# 남성 그룹의 막대를 그립니다.
# yerr 매개변수는 오류 막대를 나타내며, 평균에 대한 오차를 표시합니다.
ax.bar(ind, menMeans, width, label='Men', yerr=menStd)

# 여성 그룹의 막대를 그립니다.
# bottom 매개변수는 남성 그룹 위에 여성 그룹을 쌓도록 합니다.
# yerr 매개변수는 여성 그룹의 평균에 대한 오차를 표시합니다.
ax.bar(ind, womenMeans, width, label='Women', bottom=menMeans, yerr=womenStd)

# 그래프 제목 설정
ax.set_title('Scores by group and gender')

# 범례 추가
ax.legend()

# 그래프 출력
plt.show()
```

- np.arange(n): 0부터 n-1까지의 정수를 가지는 배열을 생성합니다.


- ax.bar(ind, menMeans, width, label='Men', yerr=menStd): 남성 그룹에 대한 막대를 그리며, yerr 매개변수는 남성 그룹의 평균에 대한 오차를 표시합니다.


- ax.bar(ind, womenMeans, width, label='Women', bottom=menMeans, yerr=womenStd): 여성 그룹에 대한 막대를 그리며, bottom 매개변수는 남성 그룹 위에 여성 그룹을 쌓도록 합니다. yerr 매개변수는 여성 그룹의 평균에 대한 오차를 표시합니다.

```python
#df = pd.read_csv()
df = pd.read_csv('./datasets/titanic.csv')
df.info()
```

```python
titanic_age = df[['Age','Survived']].dropna()
titanic_age
```

- 10세이하를 제외한 30대 이하를 기준으로 생존률 막대그래프 만들기

```python
def age_band(num):
  for i in range(1,10):
    if num < 10 * i:
      return f'under {i*10}'
titanic_age['age_band']= titanic_age['Age'].apply(age_band)
titanic_age
```


- 그룹화 및 값 계산: 'age_band'를 기준으로 데이터를 그룹화하고 'Survived' 열에 대한 값의 개수를 계산하여 'titanic_group_age'에 저장합니다.

- 결측값 처리: 'titanic_group_age'를 unstack()하여 'Survived' 값을 열로 변환하고, fillna(0)를 사용하여 결측값을 0으로 채웁니다.

- 생존율 계산: 'Survival rate' 열을 생성하여 생존율을 계산합니다. 생존율은 (생존자 수 / (생존자 수 + 비생존자 수)) * 100으로 계산됩니다.

- 결과 출력: 최종 결과를 출력합니다. 'Survival rate' 열이 추가된 DataFrame이 출력됩니다.

```python
# 'titanic_age' DataFrame에서 'age_band'와 'Survived' 열만 선택합니다.
titanic_age = titanic_age[['age_band', 'Survived']]

# 'age_band'를 기준으로 그룹화하고 'Survived' 열에 대한 값의 개수를 계산합니다.
# 그룹화된 결과를 'titanic_group_age'에 저장합니다.
titanic_group_age = titanic_age.groupby('age_band')['Survived'].value_counts().sort_index()
print(titanic_group_age)

# 'titanic_group_age'를 unstack하여 'Survived' 값을 열로 변환하고, 결측값을 0으로 채웁니다.
titanic_group_age = titanic_group_age.unstack().fillna(0)

# 생존율을 계산하여 'Survival rate' 열을 생성합니다.
# 생존율 = (생존자 수 / (생존자 수 + 비생존자 수)) * 100
titanic_group_age['Survival rate'] = (titanic_group_age[1] / (titanic_group_age[0] + titanic_group_age[1])) * 100

# 최종 결과를 출력합니다.
titanic_group_age
```

# annotate 출력하기

첫 번째 subplot: 'Age Band & Survival Rate'에 대한 막대 그래프를 그립니다. 색상은 'gray'로 설정되어 있습니다.

두 번째 subplot: 동일한 데이터에 대해 여러 스타일을 적용한 막대 그래프를 그립니다. 특정 위치의 막대는 색상을 변경하였고, 막대의 테두리 색과 두께, 투명도 등을 설정하였습니다.

막대 위에 수치 표시: 두 번째 subplot에서 각 막대 위에 해당하는 생존율을 수치로 표시하였습니다.

x축 눈금 라벨 설정: 두 번째 subplot에서 x축의 눈금 라벨을 설정하였고, 글자를 40도로 회전시켰습니다.

```python
import matplotlib.pyplot as plt

# Subplots을 생성합니다. 1행 2열의 구조로 figsize는 가로 12, 세로 5의 크기로 설정합니다.
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# 첫 번째 subplot (막대 그래프)
# 'Age Band & Survival Rate'에 대한 막대 그래프를 그립니다.
ax[0].bar(titanic_group_age.index, titanic_group_age['Survival rate'], color='gray')
ax[0].set_title('Age Band & Survival Rate')  # 그래프 제목 설정

# 두 번째 subplot (다양한 스타일 적용)
# 'Age Band & Survival Rate'에 대한 막대 그래프를 그리고 여러 스타일을 적용합니다.
color_map = ['gray'] * 9
color_map[0] = color_map[8] = '#3caea3'  # 특정 위치의 막대 색상 변경
ax[1].bar(titanic_group_age.index, titanic_group_age['Survival rate'],
          color=color_map, edgecolor='black', linewidth=1.2, alpha=0.7)  # 여러 스타일 적용
ax[1].set_title('Age Band & Survival Rate', fontsize=15, fontweight='bold', position=(0.5, 1.1))  # 그래프 제목 및 스타일 설정

# 막대 위에 수치 표시
for i, rate in enumerate(titanic_group_age['Survival rate']):
    ax[1].annotate(f'{rate:.2f}%', xy=(i, rate + 2), va='center', ha='center', fontsize=10, fontweight='bold', color='#383838') # 그림을 그리기 위함.

# x축 눈금 라벨 설정
ax[1].set_xticklabels(titanic_group_age.index, rotation=40, size=10)  # x축 눈금 라벨 설정 (글자가 겹쳐지지 않도록 회전)
plt.show()
```

# 확률분포함수

이 코드는 두 개의 정규 분포를 생성하고 그 분포들에 대한 히스토그램을 서브플롯으로 나란히 그리는 것입니다.

image.png

image.png

```python
n_point = 10000
n_bins = 20

#Generate two normal distribution
# 두 개의 정규 분포 생성
dist1 = np.random.normal(0,2,n_point)
dist2 = np.random.normal(0,10,n_point)
# dist1은 평균 (loc)이 0이고 표준 편차 (scale)가 2인 분포를 생성합니다.
# dist2는 평균 (loc)이 0이고 표준 편차 (scale)가 10인 분포를 생성합니다.

fig,axes = plt.subplots(1,2,sharey=True,tight_layout=True,figsize=(8,5))
# y축을 공유하고 레이아웃을 조정하여 서브플롯 간의 충돌을 방지
# figsize는 그림의 크기를 조절하는데 사용됨

axes[0].hist(dist1,bins=n_bins) # 데이터의 구간을 n_bins개로 나눔.
axes[1].hist(dist2, bins=n_bins)
# 둘의 차이는 스케일이 다른다. 오른쪽에 있는 그래프가 더 옆으로 퍼짐

plt.show()
```

```python
fig, axes =plt.subplots(2,1,figsize=(5,15),tight_layout=True)
axes[0].hist2d(dist1,dist2,bins=40)
axes[0].axis('equal') # axis('equal')를 안하게 되면, scale을 달리 해서 그렸는 데도 전부다 동일하게 동그랗게 그려짐
axes[1].hist2d(dist1,dist2,bins=40)

plt.show()
```

- axis('equal')를 안하게 되면, scale을 달리 해서 그렸는 데도 전부다 동일하게 동그랗게 그려짐
: 이런 경우 옆에서 보면 어떻게 보여지는지 알기 위한 코드

```python

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(projection='3d')

# 2D 히스토그램 계산
hist, xedges, yedges = np.histogram2d(dist1, dist2, bins=n_bins, range=[[-7, 7], [-30, 30]])

# 격자 생성
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing='ij')
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0
dx = dy = 0.3 * np.ones_like(zpos)
dz = hist.ravel()

# 3D 바 차트 생성
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

# 3D 시각화 설정
ax.view_init(elev=20, azim=45)

# 그래프 표시
plt.show()
```
